```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("student info.csv")
data.head()
```

```python
'''
This code converts the datatype of all the semester's SGPA from object datatype to float64.
'''
cols_to_convert = ['SEM 1 SGPA','SEM 2 SGPA','SEM 3 SGPA','SEM 4 SGPA','SEM 5 SGPA','SEM 6 SGPA','SEM 7 SGPA','SEM 8 SGPA']
for col in cols_to_convert:
    data[col] = pd.to_numeric(data[col], errors = 'coerce')
```

```python
data.dtypes
```

```python
#Filling all the missing values with median of their respective sem Sgpa's
cols_to_fill = ['SEM 1 SGPA','SEM 2 SGPA','SEM 3 SGPA']
for col in cols_to_fill:
    data[col].fillna(data[col].median(skipna = True) ,inplace = True)
```

```python
data.head(20)
```

```python
#saving as a pickle file
data.to_pickle('student_info')
```
